```python
# ðŸ”¹ Part 6: Time-Based Analysis

# Monthly revenue trend?

# Which month has highest sales?

# Daily average revenue?

import datetime as dt


df["month"] = df["order_date"].dt.month
df.groupby("month")["total_amount"].sum()


```

```python
df.groupby("order_date")["total_amount"].mean()
```

```python
# ðŸ”¹ Part 7: Visualization (Use Matplotlib Only)

# Revenue distribution (Histogram)

# Revenue by category (Bar chart)

# Monthly sales trend (Line chart)

# Experience vs Revenue type scatter (age vs total amount)

# Payment method pie chart

plt.hist(df["total_amount"] , bins=20)
plt.xlabel("total amount")
plt.ylabel("frequency")
plt.title("Revenue distribution")
plt.show()
```

```python
cat_rev = df.groupby("product_category")["total_amount"].sum()

plt.figure()
cat_rev.plot(kind="bar")
plt.title("Revenue by Category")
plt.show()
```

```python
monthely = df.groupby("month")["total_amount"].sum()

monthely.plot()
plt.title("Monthly sales trend")
plt.show()
```

```python
plt.scatter(df["age"] , df["total_amount"])
plt.xlabel("age")
plt.ylabel("total amount")
plt.title("Experience vs Revenue type")
plt.show()
```

```python
# Payment method pie chart

pay = df["payment_method"].value_counts()

plt.pie(pay, labels = pay.index)
plt.title("payment method")
plt.show()
```
